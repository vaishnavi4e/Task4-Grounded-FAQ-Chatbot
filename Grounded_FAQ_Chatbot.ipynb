```python
from google import genai
from google.colab import userdata

api_key = userdata.get("GOOGLE_API_KEY")

client = genai.Client(api_key=api_key)

print("Connected Successfully!")
```

```python
knowledge_base = """
Business Name: SkyWings Travel Agency

Business Hours:
Monday - Saturday
9:00 AM to 7:00 PM

Location:
Hyderabad, India

Contact:
Phone: +91 9876543210
Email: support@skywings.com

Services:
- Domestic Tour Packages
- International Tour Packages
- Flight Booking
- Hotel Booking
- Visa Assistance
- Travel Insurance
- Honeymoon Packages
- Group Tours
- Student Tours

FAQs

1. What are your business hours?
We are open Monday to Saturday from 9 AM to 7 PM.

2. Where are you located?
Our office is in Hyderabad.

3. Do you provide flight booking?
Yes, we book both domestic and international flights.

4. Do you provide hotel booking?
Yes, we book hotels worldwide.

5. Do you help with visas?
Yes, we provide tourist and business visa assistance.

6. Can I cancel my booking?
Yes. Cancellation charges depend on the package.

7. Do you provide refunds?
Refunds are processed according to our cancellation policy.

8. What payment methods are accepted?
UPI, Credit Card, Debit Card and Net Banking.

9. Do you offer EMI?
Yes, EMI is available on selected packages.

10. Do you provide honeymoon packages?
Yes.

11. Do you organize student tours?
Yes.

12. Do you offer travel insurance?
Yes.

13. Can I customize my trip?
Yes. We offer customized packages.

14. How do I contact customer support?
Call +91 9876543210 or email support@skywings.com

15. How early should I book?
At least 30 days before travel.
"""
```

```python
system_instruction = f"""
You are the official customer support assistant for SkyWings Travel Agency.

Business Rules:

1. Answer ONLY using the information provided in the knowledge base.
2. If the answer is not available in the knowledge base, reply exactly:
"I'm sorry, I can only answer questions related to SkyWings Travel Agency."
3. Never make up information.
4. Be polite, concise, and professional.

Knowledge Base:

{knowledge_base}
"""
```

```python
generation_config = {
    "temperature": 0.2
}

chat = client.chats.create(
    model="gemini-2.5-flash",
    history=[
        {
            "role": "user",
            "parts": [{"text": system_instruction}]
        }
    ]
)

print("✅ Chatbot Initialized Successfully!")
```

```python
print("=" * 55)
print("      Welcome to SkyWings Travel Agency")
print("=" * 55)
print("Ask your travel-related questions.")
print("Type 'quit' or 'exit' to leave.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() in ["quit", "exit"]:
        print("\nBot: Thank you for visiting SkyWings Travel Agency. Have a wonderful day!")
        break

    response = chat.send_message(
        user_input,
        config=generation_config
    )

    print("\nBot:", response.text)
    print("-" * 55)
```
